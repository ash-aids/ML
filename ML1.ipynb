```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("wine.csv")
print(data)
```

```python
X = data.drop('Customer_Segment', axis = 1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

num_components = 2
pca = PCA(n_components = num_components)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1], c = data['Customer_Segment'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Transformed Data: Wine Segment (2D)')
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
url = "Wine.csv"
df = pd.read_csv(url)

# Step 2: Preprocess the data
# The first column is the class label, so we'll separate it
y = df.iloc[:, 0]
X = df.iloc[:, 1:]

# Standardizing the data before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA
pca = PCA(n_components=2)  # We choose 2 components for visualization purposes
X_pca = pca.fit_transform(X_scaled)

# Step 4: Visualize the principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set1')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Wine Dataset')
plt.show()

# Step 5: Inspect the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by each component: {explained_variance}")
print(f"Total explained variance by 2 components: {np.sum(explained_variance)}")
```
