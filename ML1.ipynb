```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("wine.csv")
print(data)
```

```python
X = data.drop('Customer_Segment', axis = 1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

num_components = 2
pca = PCA(n_components = num_components)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1], c = data['Customer_Segment'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Transformed Data: Wine Segment (2D)')
plt.show()
```

```python
# Explained Variance (to understand how much variance is captured by the components)
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by components: {explained_variance}")
print(f"Total explained variance: {explained_variance.sum()}")
```
